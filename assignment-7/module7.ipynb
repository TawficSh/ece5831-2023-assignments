```python
import random
from sklearn.utils import shuffle
#from mnist_keras import load_data, load, test, predict
import numpy as np
import matplotlib.pyplot as plt

def display_images(images, titles, rows, cols):
    for i in range(len(images)):
        plt.subplot(rows, cols, i+1)
        plt.imshow(images[i], cmap='gray')
        plt.title(titles[i])
        plt.axis('off')
    plt.show()

if __name__ == '__main__':
    (_, _), (x_test, y_test) = m.load_data()
    x_test, y_test = shuffle(x_test, y_test)
    model = m.load("model_your_Tawfic_Shamieh")
    predictions = model.predict(x_test)
    unsuccessful_indices = [i for i in range(len(y_test)) if np.argmax(predictions[i]) != y_test[i]]

    for _ in range(2):
        selected_indices = random.sample(unsuccessful_indices, 12)
        images = [x_test[i].reshape(28, 28) for i in selected_indices]
        titles = [f"True: {y_test[i]}, Pred: {np.argmax(predictions[i])}" for i in selected_indices]
        display_images(images, titles, 3, 4)
```
